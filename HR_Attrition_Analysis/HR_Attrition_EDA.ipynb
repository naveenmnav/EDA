# **1. Introduction**
---


- Attrition is a big concern for IT Companies. The term **Attrition** refers to a gradual but deliberate reduction in staff numbers that occurs as employees retire or resign and are not replaced.


- Attrition happens for several reasons, including pay, lack of growth, and poor workplace conditions. The term is also sometimes used to describe the loss of customers or clients as they mature beyond a product or company's target market without being replaced by a younger generation.


- Industry analysts anticipate a 22-23% attrition rate in 2021, which works out to one million resignations on a projected base of 4.6 million IT employees.


- This analysis is done based on the Attrition Dataset of an IT Firm based out of India

<a name = Section51></a>
### **5.1 Handling of Missing Data**

- In this section, we will identify missing data and check the proportion of it and take appropriate measures.

```python
missing_data = pd.DataFrame( data = dataset.shape[0] - dataset.count(), columns = ['Missing Features Count'])

missing_data['Missing Feature %'] = 100 * (missing_data['Missing Features Count'] / dataset.shape[0])
```

```python
missing_data
```

**Observations:**

- We can observe that following features are found to have missing values along with the proportions:

|Feature|Object Type|Missing Proportion|Solution|
|:--:|:--:|:--:|:--|
|PercentSalaryHike	|float64|0.067%|Replace with mean.|
|TrainingTimesLastYear|int64|1.561%|Replace with median.|
|Mode_of_work|Object|0.067%|Replace with mode.|
|YearsSinceLastPromotion|float64|0.067%|Replace with median.|

**Performing Operations**

```python
# Filling up the missing values



# 1. Filling the missing PercentSalaryHike with the mean
dataset['PercentSalaryHike'] = dataset['PercentSalaryHike'].replace(np.nan, dataset['PercentSalaryHike'].mean())

#2. Filling the missing TrainingTimesLastYear with median
dataset['TrainingTimesLastYear'] = dataset['TrainingTimesLastYear'].replace(np.nan, dataset['TrainingTimesLastYear'].median())

#3. Filling the missing Mode_of_work with the mode
dataset['Mode_of_work'] = dataset['Mode_of_work'].replace(np.nan, dataset['Mode_of_work'].mode()[0])


#4. Filling the YearsSinceLastPromotion with the median
dataset['YearsSinceLastPromotion'] = dataset['YearsSinceLastPromotion'].replace(np.nan, dataset['YearsSinceLastPromotion'].median())
```

**Let us verify the integrity of the null values**

```python
null_dataset = pd.DataFrame(index = dataset.columns.values)
null_dataset['Null Frequecy'] = dataset.isnull().sum().values
null_dataset['Null Percentage'] = 100 * null_dataset['Null Frequecy'] / dataset.shape[0]


null_dataset
```

### **5.3 Handling of Outliers**

- In this section, we will identify **outliers** in data and and then **take appropriate measures**.

- The **MonthlyIncome** has a standard deviation of **17750**. We need to analyse further on this.

- As we don't know what had caused this deviation, we can plot a graph and observe

```python
fig = plt.figure(figsize=(15, 7))
plt.plot(dataset['MonthlyIncome'])
plt.xlabel('Employee Index', size = 14)
plt.ylabel('Monthly Salary', size = 14)
plt.xticks(size=12)
plt.yticks(size=12)
plt.grid(b = True)
plt.show()
```

**Observations**

- Here we can see there are 2 spikes representing outliers

- Let us analyse and decide if we need to remove the outiers or make it consistent


```python
# Taking the max value between 0 to 200 employees

print(dataset['MonthlyIncome'][:200].max())

#Taking the max value between 1400 to 1473 employees
print(dataset['MonthlyIncome'][1400:].max())
```

```python
dataset['MonthlyIncome'].mean()
```

**Observations**

- We can see the Maximum Monthly Income is Rs. 85447, which can be considered as normal for an employee who is in a Management role

- Age has a record of **-122**. As per the rules, the minimum age should be 21. So we can change the value to 21

- But the Average Monthly Income is Rs. 7001. 

- We infer there are outliers in the dataset.

- From the ablove plot, we can find that there 2 records which are potential outliers. 

- We can remove these outliers as there is no evidence to alter the values with standard values

```python
#Removing the outliers in MonthlySalary

dataset.loc[dataset['MonthlyIncome']==85447]
dataset.drop(79, axis = 0, inplace = True)    
dataset.loc[dataset['MonthlyIncome']==658699]
dataset.drop(1471, axis = 0, inplace = True)
```

```python
# Removing outlier in Age

dataset['Age'][dataset['Age'] < 21] = 21
```
